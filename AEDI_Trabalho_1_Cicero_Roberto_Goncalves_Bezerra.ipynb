### **5. Fundamentação Teórica**

####**5.1. Distribuição de Poisson**####


**Definição**
> Seja uma variável aleatória discreta, tomando os seguintes valores: $0, 1, \dots, n, \dots$ Se
\begin{align}
        P(X=k)=\dfrac{e^{-\lambda}\lambda^k}{k!},\ \ \ \ \ k=0,1,\dots,n,\dots,
    \end{align}
> podemos dizer que $X$ tem *distribuicao de Poisson*, com parâmetro $\lambda>0$.

Observe que há representação legítima da distribuição de probabilidade através de
\begin{align}
        \sum_{k=0}^{\infty}P(X=k)=\sum_{k=0}^{\infty}\left(\dfrac{e^{-\lambda}\lambda^k}{k!}\right)=e^{-\lambda}e^{\lambda}=1.
    \end{align}

**Comentário**
> Podemos supor que o espaço amostral $S$ tenha sido identificado com $R_X$ e que $X(s)=s$, uma vez em que estamos definindo a variável aleatória diretamente em termos de seu contradomínio e distribuição de probabilidade, sem referência a nenhum espaço amostral subjacente $S$.
> Ou seja, os resultados do experimento são apenas os números $0,1,2,\dots$ e as probabilidades associadas a cada um desses resultados são dadas por
>\begin{align}
        P(X=k)=\dfrac{e^{-\lambda}\lambda^k}{k!},\ \ \ \ \ k=0,1,\dots,n,\dots,
    \end{align}

**Teorema**
> Se X tiver distribuição de Poisson com parâmetro $\lambda$, então
> \begin{align}
          E(X) = \lambda\ \ \  \text{ e }\ \ \ V(X) = \lambda.
    \end{align}

**Demonstração**
\begin{align}
        E(X)=\sum_{k=0}^{\infty}\dfrac{ke^{-\lambda}\lambda^k}{k!}=\sum_{k=1}^{\infty}\dfrac{e^{-\lambda}\lambda^k}{(k-1)!}.
    \end{align}

Fazendo-se $s = k – 1$, temos o seguinte
\begin{align}
        E(X)=\sum_{s=0}^{\infty}\dfrac{e^{-\lambda}\lambda^{s+1}}{s!}=\lambda\sum_{s=0}^{\infty}\dfrac{e^{-\lambda}\lambda^s}{s!}=\lambda.
    \end{align}


Da mesma forma,
\begin{align}
        E(X^2)=\sum_{k=0}^{\infty}\dfrac{k^2e^{-\lambda}\lambda^{k}}{k!}=\sum_{k=1}^{\infty}k\dfrac{e^{-\lambda}\lambda^k}{(k-1)!}.
    \end{align}


Efeuando o mesmo procedimento, ou seja, $s=k-1$, temos que
\begin{align}
        E(X^2)=\sum_{s=0}^{\infty}(s+1)\dfrac{e^{-\lambda}\lambda^{s+1}}{s!}=\lambda\sum_{s=0}^{\infty}s\dfrac{e^{-\lambda}\lambda^s}{s!}+\lambda\sum_{s=0}^{\infty}\dfrac{e^{-\lambda}\lambda^s}{s!}=\lambda^2 + \lambda
    \end{align}

porque o primeiro somatório representa E(X), enquanto o segundo é igual a um. Portanto,
\begin{align}
        V(X) = E(X^2) – [E(X)]^2 = \lambda^2 + \lambda – \lambda^2 = \lambda.
    \end{align}


Ou seja, uma variável aleatória de Poisson possui essa propriedade notável: o seu valor esperado é igual à sua variância. Isso implica que, ao aumentar a taxa média de eventos, tanto a média quando a variância da distribuição também aumentam.

####**5.2. A Distribuição de Poisson como aproximação da Distribuição Binomial**####


A distribuição de Poisson representa um papel muito importante como um modelo probabilístico adequado para um grande número de fenômenos aleatórios.
Para definir a distribuição de Poisson, observamos o que acontece com as probabilidades binomiais $\binom{n}{k}p^k (1-p)^{n-k}$, se $n\rightarrow \infty$ e $p \rightarrow 0$, de forma que $np$ permaneça constante, digamos $np=\lambda$.

Considerando a expressão geral da probabilidade binomial,
>\begin{align}
        P(X=k)&=\binom{n}{k}p^k (1-p)^{n-k}= \\
        &=\dfrac{n!}{k!(n-k)!}p^k (1-p)^{n-k}= \\
        &=\dfrac{n(n-1)(n-2)\dots(n-k+1)}{k!}p^k (1-p)^{n-k}.
    \end{align}

Considerando $np=\lambda$, temos que $p=\lambda/n$, e $1-p=1-\lambda/n=(n-\lambda)/n$. Ao substituir todos os termos que contenham $p$ pelas expressões equivalentes com $\lambda$, temos

>\begin{align}
        P(X=k)&=\dfrac{n(n-1)\dots(n-k+1)}{k!}\left(\dfrac{\lambda}{n}\right)^k \left(\dfrac{n-\lambda}{n}\right)^{n-k} = \\
        &=\dfrac{\lambda!}{k!}\left[(1)\left(1-\dfrac{1}{n}\right)\left(1-\dfrac{2}{n}\right)\dots\left(1-\dfrac{k-1}{n}\right)\right]\left[1-\dfrac{\lambda}{n}\right]^{n-k}= \\
        &=\dfrac{\lambda!}{k!}\left[(1)\left(1-\dfrac{1}{n}\right)\left(1-\dfrac{2}{n}\right)\dots\left(1-\dfrac{k-1}{n}\right)\right]\times \left[1-\dfrac{\lambda}{n}\right]^{n}\left[1-\dfrac{\lambda}{n}\right]^{-k}.
    \end{align}

Dessa vez, vamos fazer $n\rightarrow \infty$, de tal modo que $np=\lambda$ permaneça constante. Isto significa que $p\rightarrow 0$, enquanto $n\rightarrow\infty$.

Verificamos na expressão que os termos da forma $(1-1/n)$, $(1-2/n)$, $\dots$ se aproximam da unidade à medida que $n$ tende para o infinito, gerando $(1-\lambda/n)^{-k}$.

Segundo a definição do número $e$, temos que $(1-\lambda/n)^n \rightarrow e^{-\lambda}$ quando $n \rightarrow \infty$.

Então, $\lim_{n\rightarrow\infty}P(X=k)=e^{-\lambda} \lambda^{k}/k!$. Isso traz a distribuição de Poisson com o parâmetro $\lambda$.

> **Teorema**
> Seja $X$ uma variável aleatória distribuída binomialmente com parâmetro $p$ (baseado em $n$ repetições de um experimento). Isto é,
>\begin{align}
        P(X=k)&=\binom{n}{k}p^k (1-p)^{n-k}.
    \end{align}
> Admita-se que quando $n\rightarrow\infty$, fique $np=\lambda$(constante), ou equivalentemente, quando $n\rightarrow\infty$, $p\rightarrow 0$, de modo que $np\rightarrow\lambda$. Nessas condições teremos
>\begin{align}
        \lim_{n\rightarrow\infty}. P(X=k)=\dfrac{e^{-\lambda}\lambda^{k}}{k!}
    \end{align},
> que é a distribuição de Poisson com o parâmetro $\lambda$.

**Comentários**

(a) Poderemos obter uma aproximação das probabilidades binomiais com as probabilidades da distribuição de Poisson, sempre que o $n$ seja grande e $p$ seja pequeno.

(b) Se $X$ tiver uma distribuição binomial, temos que $E(X)=np$, enquanto que, se $X$ tiver uma distribuição de Poisson (com parâmetro $\lambda$), temos que $E(X)=\lambda$.

(c) A distruibuição binomial é caracterizada por dois parâmetros: $n$ e $p$, enquanto que a distribuição de Poisson é caracterizada por um único parâmetro: $\lambda=np$. O $\lambda$ representa o número esperado de sucessos por unidade de tempo. Esse parâmetro também é conhecido por 'intensidade' da distribuição.

(d) Para avaliar a variância de uma variável aleatória de Poisson $X$, com parâmetro $\lambda$, $X$ pode ser considerado como um caso-limite de uma variável aleatória distribuída binomialmente $Y$, com parâmetros $n$ e $p$, na qual $n\rightarrow\infty$ e $p\rightarrow 0$, de modo que $np\rightarrow\lambda$. Desde que $E(Y)=np$ e $Var(Y)=np(1-p)$, observamos que, no limite, $Var(Y)\rightarrow\lambda$.

####**5.3 Comparação com outras distribuições**
**Distribuição Binomial**

A Distribuição de Poisson pode ser entendida como um caso limite da Distribuição Binomial. Quando o número de tentativas $n$ é grande e a probabilidade de sucesso $p$ é pequena, mantendo $\lambda=np$ constante, a Distribuição Binomial converge para uma Distribuição de Poisson.

**Distribuição Normal**

Para valores elevados de $\lambda$, a Distribuição de Poisson tende a aproximar-se de uma Distribuição Normal com média $\lambda$ e variância $\lambda$. Essa aproximação ocorre devido ao Teorema Central do Limite, que estabelece que a soma de um grande número de variáveis aleatórias independentes e identicamente distribuídas tende a seguir uma distribuição normal, independentemente da distribuição original dessas variáveis.

No contexto da Distribuição de Poisson, isso significa que, para $\lambda$ suficientemente grande, os valores da variável aleatória podem ser tratados como aproximadamente normais, facilitando cálculos e análises estatísticas em casos onde $\lambda$ assume valores altos.

####**5.4 Métodos de Monte Carlo**

O Método de Monte Carlo (MMC) refere-se a uma classe de métodos estatísticos que utiliza amostragens aleatórias em grande escala para obter resultados numéricos. Em resumo, esses métodos empregam a aleatoriedade dos dados para gerar soluções para problemas que, a princípio, seriam determinísticos. São amplamente aplicados em problemas de física e matemática que apresentam alto grau de complexidade ou são inviáveis de resolver por outros métodos.

A simulação de variáveis aleatórias deu origem aos métodos Monte Carlo (MMC), que partem da premissa de que o pesquisador tem acesso a um gerador de números aleatórios equiprováveis. Um número aleatório (NA) representa o valor de uma variável aleatória distribuída uniformemente no intervalo $(0, 1)$. Originalmente, esses números eram gerados manualmente ou de forma mecânica, com o uso de dados, roletas e outros dispositivos. Atualmente, utilizamos computadores para gerar números que, embora não sejam verdadeiramente aleatórios, são pseudo-aleatórios.

Para nossos objetivos, podemos entender uma simulação como uma realização específica de um modelo (como o binomial, normal etc.). Assim, os valores simulados podem ser considerados uma amostra, como veremos nos próximos capítulos. Esse entendimento é útil para analisar as distribuições de estimadores e suas propriedades.

O nome Monte Carlo tem origem na cidade homônima, no Principado de Mônaco, devido à famosa roleta, um exemplo de mecanismo para gerar números aleatórios. Os MMC surgiram durante a Segunda Guerra Mundial em pesquisas sobre a difusão aleatória de nêutrons em materiais radioativos. Esse método foi inicialmente desenvolvido por Ulam, Metropolis, Fermi e von Neumann, por volta de 1948-1949. Trabalhos de referência sobre o tema incluem os artigos de Metropolis e Ulam (1949), von Neumann (1951) e os livros de Sóbol (1976), Hammersley e Handscomb (1964) e Ross (1997).

### **6. Construção do Modelo**

O código a seguir representa uma análise de uma Infraestrutura de TI com 10000 servidores e que possui uma quantidade esperada de falhas mensais, seja por problema inerente ao Sistema Operacional ou em decorrência de falha humana, como por exemplo, a alteração indevida de sudoers, a mudança em permissionamento e outros que podem gerar falhas de segurança.

O ambiente é monitorado por ferramenta de gerenciamento de desempenho e análise de operações de TI que gera a estimação de falhas mensais.

Serão efetuadas simulações de Monte Carlo para a obtenção de aproximações que auxiliarão na análise do cenário.

Para implementar a simulação e estimar o comportamento da infraestrutura de TI, utilizaremos as seguintes variáveis principais:

*   **Número de Servidores (n_servers):** Total de servidores na infraestrutura de TI simulada.
*   **Taxa Média de Falhas (λ_failures):** Número médio de falhas que um servidor experimenta por unidade de tempo (ex.: falhas por mês).
*   **Tempo Médio de Reparo (λ_repair):** Tempo médio necessário para recuperar um servidor após uma falha, normalmente medido em horas.
*   **Período de Simulação (months_in_a_year):** Definido como um ano (12 meses), dividido em períodos de um mês para a simulação das falhas.
*   **Número de Simulações (simulations):** Total de iterações da simulação de Monte Carlo, que será utilizado para capturar uma gama variada de possíveis cenários.

Segue abaixo o código do modelo:

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros de entrada
lambda_failures = 0.00167  # Média de falhas esperadas por mês
lambda_repair = 1 / 5  # Taxa de reparo: em média 5 horas para reparar
n_servers = 10000
months_in_a_year = 12
simulations = 1000  # Número de simulações de Monte Carlo

# Função para simular um ano de operação com falhas e reparos
def simulate_downtime_poisson(lambda_failures, lambda_repair, n_servers, months_in_a_year):
    monthly_downtime = []
    for _ in range(n_servers):
        server_monthly_downtime = []
        for _ in range(months_in_a_year):
            failures = np.random.poisson(lambda_failures)  # Número de falhas no mês
            downtime = sum(np.random.exponential(1 / lambda_repair, failures))  # Tempo total de inatividade no mês
            server_monthly_downtime.append(downtime)
        monthly_downtime.append(server_monthly_downtime)
    return np.sum(monthly_downtime, axis=0)  # Downtime total mensal de todos os servidores

# Executar várias simulações de Monte Carlo
downtime_results = [simulate_downtime_poisson(lambda_failures, lambda_repair, n_servers, months_in_a_year) for _ in range(simulations)]

# Converter resultados em array para facilitar as análises
downtime_results = np.array(downtime_results)
total_downtime_annual = downtime_results.sum(axis=1)  # Inatividade total anual em cada simulação
mean_monthly_downtime = downtime_results.mean(axis=0)  # Inatividade média mensal entre as simulações
availability = 1 - (total_downtime_annual / (months_in_a_year * 30 * 24 * n_servers))  # Disponibilidade anual

# Estatísticas descritivas
mean_downtime = np.mean(downtime_results)
mean_availability = 1 - (mean_downtime / (months_in_a_year * 30 * 24 * n_servers))  # 30 dias * 24 horas

print(f"Tempo médio de inatividade (horas/ano): {mean_downtime:.5f}")
print(f"Disponibilidade média do sistema: {mean_availability * 100:.5f}%")

# 1. Histograma do Tempo Total de Inatividade Anual
plt.figure(figsize=(10, 6))
plt.hist(total_downtime_annual, bins=50, color='skyblue', alpha=0.7, edgecolor='black')
plt.title('Histograma do Tempo Total de Inatividade Anual')
plt.xlabel('Tempo de Inatividade (horas)')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. Gráfico de Linhas do Tempo Médio de Inatividade por Mês
plt.figure(figsize=(10, 6))
plt.plot(range(1, months_in_a_year + 1), mean_monthly_downtime, marker='o', color='blue')
plt.title('Tempo Médio de Inatividade por Mês')
plt.xlabel('Mês')
plt.ylabel('Tempo Médio de Inatividade (horas)')
plt.grid(True)
plt.show()

# 3. Boxplot da Disponibilidade Anual
plt.figure(figsize=(8, 6))
plt.boxplot(availability * 100, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Distribuição da Disponibilidade Anual do Sistema')
plt.xlabel('Disponibilidade (%)')
plt.show()
```

### **7. Resultados da Simulação**

Foram realizadas 1000 simulações de Monte Carlo para cada servidor, onde foram simuladas falhas e reparos durante o período de um ano, utilizando a função $simulate\_downtime\_poisson$.

Para cada servidor foram calculados o número de falhas no mês com o $np.random.poisson$, tendo como parâmetro a média de falhas esperadas pelo servidor a cada mês. Em seguida  calculado o tempo total de inatividade do servidor no mês com o somatório dos resultados da função $np.random.exponencial$, tendo como parâmetros a taxa de reparo e o número de falhas por mês.

A análise reforça a importância da escolha de uma distribuição apropriada para simulações complexas e ressalta o valor do planejamento estatístico para a gestão de grandes infraestruturas de TI.
